# Quantum Feature Map

In machine learning, for example in SVM, it is usual to transform the data from a space of low dimension to a higher dimension, for example 2D --> 3D; this is made because sometimes the data it is not linearly separable.
Thhat transformation looks like

$\phi: X \rightarrow F$

Where $F$ is called the feature space, and the points are called the feature vectors: $\phi(x) \  \forall x \  \in \ X$
